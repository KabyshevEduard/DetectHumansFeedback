## Preprocessed

```python
df = pd.read_csv(path + 'preprocessed_df.csv')
df.head()
```

```python
nltk.download('stopwords')
```

```python
def postpreprocess(text, lng: str):
    try:
        text = text.split(' ')
        stops = set(nltk.corpus.stopwords.words(lng))
        filtered = [word for word in text if word not in stops]
        stemmer = nltk.stem.SnowballStemmer(lng)
        filtered = [stemmer.stem(word) for word in filtered]
        text = ' '.join(filtered)
        return text
    except:
        return np.NaN
```

```python
df['text'] = df.apply(lambda row: postpreprocess(row['text'], 'english'), axis=1)
```

```python
df = df.dropna()
```

```python
df.to_csv(path + 'preprocessed_df(stemmed + unstopped).csv')
```
