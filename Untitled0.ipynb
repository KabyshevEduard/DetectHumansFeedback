```python
#torch custom dataset processor
class CustomTextDataset(torch.utils.data.Dataset):
    def __init__(self, df_x, df_label, text_field, tokenizer="google-bert/bert-base-uncased"):
        self.df_x = df_x
        self.df_label = df_label
        self.text_field = text_field
        self.tokenizer = AutoTokenizer.from_pretrained(tokenizer)

    def __len__(self):
        return len(self.df_x)

    def __getitem__(self, idx):
        text = self.df_x[self.text_field].iloc[idx]
        tokenized_text = self.tokenizer(text, padding="max_length", truncation=True, return_tensors='pt')
        for key in tokenized_text.keys():
            tokenized_text[key] = tokenized_text[key].squeeze(0)
        fields = [column for column in list(self.df_x.columns) if (column != self.text_field)]
        others = self.df_x[fields].iloc[idx].values
        others = torch.from_numpy(others)
        others = others.to(torch.float32)
        label = self.df_label.iloc[idx]
        label = torch.tensor([label], dtype=torch.float32)
        return tokenized_text, others, label
```

```python
train_data = CustomTextDataset(X_train, y_train, 'text')
test_data = CustomTextDataset(X_test, y_test, 'text')
dataloader_train = torch.utils.data.DataLoader(train_data, batch_size=196, shuffle=True)
dataloader_test = torch.utils.data.DataLoader(test_data, batch_size=196, shuffle=True)
```

```python
#our model from picture
class CustomBERTModel(torch.nn.Module):
    def __init__(self, m, bert_model, num_classes, bert_output_size=(512, 768)):
        super().__init__()
        self.bert = bert_model
        self.classifier = torch.nn.Linear(bert_output_size[0]*bert_output_size[1] + m, num_classes, dtype=torch.float32)
        self.flat = torch.nn.Flatten()

    def forward(self, x_bert, x_one_hot_table):
        outputs_bert = self.bert(**x_bert)
        x_flatten = self.flat(outputs_bert.last_hidden_state)
        x = torch.cat((x_flatten, x_one_hot_table), -1)
        x = self.classifier(x)
        y = torch.nn.functional.sigmoid(x)
        return y
```

```python
from tqdm.auto import tqdm
import IPython
from torcheval.metrics import BinaryPrecision, BinaryRecall
```

```python
#train class
class Trainer:
    n = 0

    def __init__(self, full_model):
        self.full_model = full_model
        self.__epoches_values = []
        self.__x_epoches_values = []
        self.__batches_values = []
        self.__x_batches_values = []

    def __progress(self, value, epoch, pres, rec):
        return IPython.display.HTML("""
                <p>Epoch {epoch}: loss = {value}; precision = {pres}, recall = {rec}</p>
            """.format(value=value, epoch=epoch, pres=pres, rec=rec))

    def fit_model(self, loss_function, num_epochs, train_loader, test_loader, device):
        progress_bar_epoches = tqdm(range(num_epochs))
        optimizer = torch.optim.AdamW(self.full_model.parameters(), lr=0.0001, weight_decay=0.5)
        mean_loss_per_epoch = 0
        prec = 0
        rec = 0
        for epoch in range(1, num_epochs+1):
            progress_bar_epoches.update(1)
            progress_bar_batches = tqdm(range(len(train_loader)))
            out = display(self.__progress(mean_loss_per_epoch, epoch, 0, 0), display_id=True)
            self.full_model.train()
            for i, (x_bert, x_one_hot_col, label) in enumerate(train_loader):
                progress_bar_batches.update(1)
                optimizer.zero_grad()
                x_bert = x_bert.to(device)
                x_one_hot_col = x_one_hot_col.to(device)
                label = label.to(device)
                output = self.full_model(x_bert, x_one_hot_col)
                loss = loss_function(output, label)
                loss.backward()
                optimizer.step()
                self.n += 1
                mean_loss_per_epoch += float(loss)
                mean_loss_per_epoch = mean_loss_per_epoch/self.n

                self.__batches_values.append(mean_loss_per_epoch)
                self.__x_batches_values.append(self.n)
                out.update(self.__progress(mean_loss_per_epoch, epoch, prec, rec))

            prec, rec = self.__calculate_metric(test_loader)
            out.update(self.__progress(mean_loss_per_epoch, epoch, prec, rec))
            progress_bar_batches.close()
            self.__epoches_values.append(mean_loss_per_epoch)
            self.__x_epoches_values.append(epoch)

    def get_learning_line_epoches(self):
        y = np.array(self.__epoches_values)
        x = np.array(self.__x_epoches_values)
        plt.plot(x, y)
        plt.ylabel('Loss')
        plt.xlabel('epoch')
        plt.show()

    def get_learning_line_batches(self):
        y = np.array(self.__batches_values)
        x = np.array(self.__x_batches_values)
        plt.plot(x, y)
        plt.ylabel('Loss')
        plt.xlabel('batch')
        plt.show()

    def get_model(self):
        return self.full_model

    def __calculate_metric(self, loader):
        precision_func = BinaryPrecision()
        recall_func = BinaryRecall()
        self.full_model.eval()
        with torch.no_grad():
            for x, x_1, y in loader:
                x, x_1, y = x.to(device), x_1.to(device), y.to(device)
                targets = torch.reshape(y, (-1, )).to(dtype=torch.int8)
                preds = torch.round(self.full_model(x, x_1))
                preds = torch.reshape(preds, (-1, )).to(dtype=torch.int8)
                precision_func.update(preds, targets)
                recall_func.update(preds, targets)
        return float(precision_func.compute()), float(recall_func.compute())
```

```python
def get_metrics(loader, model, device):
    model = model.to(device)
    precision_func = BinaryPrecision()
    recall_func = BinaryRecall()
    for x, x_1, y in loader:
        x, x_1, y = x.to(device), x_1.to(device), y.to(device)
        targets = torch.reshape(y, (-1, )).to(dtype=torch.int8)
        preds = torch.round(model(x, x_1))
        preds = torch.reshape(preds, (-1, )).to(dtype=torch.int8)
        precision_func.update(preds, targets)
        recall_func.update(preds, targets)
    return {'precision': float(precision_func.compute()), 'recall': float(recall_func.compute())}
```

```python
class LSTM_model(torch.nn.Module):
    def __init__(self, m, num_classes, max_length=512):
        super().__init__()
        self.emb = torch.nn.Embedding(30522, max_length)
        self.lstm = torch.nn.LSTM(max_length, max_length, 1)
        self.drop1 = torch.nn.Dropout(p=0.2)
        self.classifier = torch.nn.Linear(262156, num_classes, dtype=torch.float32)
        self.drop2 = torch.nn.Dropout(p=0.2)
        self.flat = torch.nn.Flatten()

    def forward(self, tokens, x_one_hot_table):
        ids = tokens['input_ids']
        x = self.emb(ids)
        y, h = self.lstm(x)
        y = self.flat(y)
        y = self.drop1(y)
        z = torch.cat((y, x_one_hot_table), -1)
        z = self.classifier(z)
        z = self.drop2(z)
        z = torch.nn.functional.sigmoid(z)
        return z
```
